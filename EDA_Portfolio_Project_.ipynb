```python
# Statistical Summary Part 2 (Categorical Data)

cat_summary = df.describe(include=['object'])
print(cat_summary)
```

```python
# Non-Graphical Analysis
## Value Counts for all categorical features

## We have three categorical variables - Product, Gender, Marital Status


for col in df.select_dtypes(include=['object']).columns:  # Select categorical columns
    print(f"\nValue Counts for {col}:\n")
    print(df[col].value_counts())
```

```python
# Graphical Analysis Univariate
# Numerical (Distribution Plot)


sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title("Distribution Plot of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Graphical Analysis Univariate
# Numerical (Distribution Plot)


sns.histplot(df['Education'], kde=True, bins=30, color='skyblue')
plt.title("Distribution Plot of Education")
plt.xlabel("Education")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df['Usage'], kde=True, bins=30, color='skyblue')
plt.title("Distribution Plot of Usage")
plt.xlabel("Usage")
plt.ylabel("Frequency")
plt.show()
```

```python
# Graphical Analysis Univariate
# Categorical (Count Plot)


sns.countplot(x='Product', data=df, palette='Set2')
plt.title("Count Plot of Products")
plt.xlabel("Product")
plt.ylabel("Count")
plt.show()
```

```python
# Graphical Analysis Univariate
# Categorical (Count Plot)



sns.countplot(x='Gender', data=df, palette='Set2')
plt.title("Count Plot of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Graphical Analysis Univariate
# Categorical (Count Plot)



sns.countplot(x='MaritalStatus', data=df, palette='Set2')
plt.title("Count Plot of Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
```

```python
# Bivariate Analysis
# Product Vs Gender

plt.figure(figsize=(6, 4))
sns.countplot(x='Product', hue='Gender', data=df, palette='Set2')
plt.title("Product Purchased by Gender")
plt.xlabel("Product")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

```python
# Bivariate Analysis
# Product Vs Marital Status

plt.figure(figsize=(6, 4))
sns.countplot(x='Product', hue='MaritalStatus', data=df, palette='Set2')
plt.title("Product Preferences by Marital Status")
plt.xlabel("Product")
plt.ylabel("Count")
plt.legend(title="MaritalStatus")
plt.show()
```

```python
# Bivariate Analysis
# Product Vs Age


bins = [0, 18, 30, 45, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

plt.figure(figsize=(8, 5))
sns.countplot(x='Product', hue='Age_Group', data=df, palette='Set2')
plt.title("Product Purchases by Age Group")
plt.xlabel("Product")
plt.ylabel("Count")
plt.legend(title="Age Group")
plt.show()
```

```python
# Multivariate Analysis

sns.pairplot(df)
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
#Correlation Matrix

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Conditional Probabilities

print(df['Product'].value_counts(normalize=True) * 100)

print('\nProduct vs Gender:')
print(pd.crosstab(df['Product'], df['Gender'], normalize='columns') * 100)

female_kp781 = len(df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')])
female_total = len(df[df['Gender'] == 'Female'])
print('% Female buying KP781:', (female_kp781 / female_total) * 100)

female_given_kp281 = df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')]
print('P(Female | KP281):', len(female_given_kp281) / len(df[df['Product'] == 'KP281']))

age_20_30 = df[(df['Age'] >= 20) & (df['Age'] <= 39)]
print('% customers aged 20–39:', len(age_20_30) / len(df) * 100)

low_income = df['Income'] < 40000
high_income = df['Income'] > 80000
print('% Low-income buying:', df[low_income].shape[0] / df.shape[0] * 100)
print('% High-income buying KP781:', df[(high_income) & (df['Product'] == 'KP781')].shape[0] / df.shape[0] * 100)
print('P(High-income | KP781):', df[(high_income) & (df['Product'] == 'KP781')].shape[0] / df[df['Product'] == 'KP781'].shape[0])

fit5 = df[df['Fitness'] == 5]
print('% with Fitness level 5:', fit5.shape[0] / df.shape[0] * 100)
print('% of fit5 buying KP781:', fit5[fit5['Product'] == 'KP781'].shape[0] / fit5.shape[0] * 100)
print('P(fit5 | KP781):', fit5[fit5['Product'] == 'KP781'].shape[0] / df[df['Product'] == 'KP781'].shape[0])

print('% Partnered customers:', df[df['MaritalStatus'] == 'Partnered'].shape[0] / df.shape[0] * 100)
```
