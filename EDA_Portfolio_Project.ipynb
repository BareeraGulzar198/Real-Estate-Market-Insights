```python
def convert_price(value):
    if not isinstance(value, str):
        return None

    # Clean text
    value = value.replace(',', '').strip().lower()

    # Conversion logic
    try:
        if 'crore' in value:
            num = float(value.replace('crore', '').strip())
            return num * 10000000

        elif 'lakh' in value:
            num = float(value.replace('lakh', '').strip())
            return num * 100000

        elif 'thousand' in value:
            num = float(value.replace('thousand', '').strip())
            return num * 1000

        else:
            # if value is already numeric
            return float(value)
    except:
        return None


df_cleaned['Price_PKR'] = df_cleaned['Price'].apply(convert_price)

print(df_cleaned['Price_PKR'])


```

```python
df_cleaned["Area"].unique()
```

```python
df_cleaned['Area'] = df_cleaned['Area'].str.replace(',', '', regex=True)



def convert_area(value):
    if not isinstance(value, str):
        return None

    # Step 1: Clean text
    value = value.replace(',', '')        # remove commas
    value = value.replace('.', '')        # remove periods (from 'Sq. Yd.')
    value = value.strip().lower()

    # Step 2: Conversion logic
    if 'sq yd' in value:
        # Remove text and convert directly to float
        num = value.replace('sq yd', '').strip()
        return float(num)

    elif 'marla' in value:
        num = value.replace('marla', '').strip()
        return float(num) * 30.25

    elif 'kanal' in value:
        num = value.replace('kanal', '').strip()
        return float(num) * 605

    else:
        return None  # unknown format

# Apply conversion

df_cleaned['Area_SqYd'] = df_cleaned['Area'].apply(convert_area).astype(float)

print(df_cleaned['Area_SqYd'])


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ['Parking Spaces', 'Built in year', 'Store Rooms', 'Kitchens', 'Servant Quarters']

# Make a copy before outlier removal
df_before = df.copy()

# --- Outlier removal ---
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

# --- Visual comparison ---
fig, axes = plt.subplots(nrows=len(numeric_cols), ncols=2, figsize=(12, 4 * len(numeric_cols)))
fig.suptitle("Before vs After Outlier Removal (IQR Method)", fontsize=16)

for i, col in enumerate(numeric_cols):
    sns.boxplot(x=df_before[col], ax=axes[i, 0], color="salmon")
    axes[i, 0].set_title(f"{col} â€” Before Cleaning")

    sns.boxplot(x=df[col], ax=axes[i, 1], color="lightgreen")
    axes[i, 1].set_title(f"{col} â€” After Cleaning")

plt.tight_layout(pad=3.0)
plt.show()
```
